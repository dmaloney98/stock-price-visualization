```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Replace the path with the actual path to the GOOGL.csv file on your desktop
data = pd.read_csv("/Users/donaldmaloney/Desktop/GOOGL.csv")
```

```python
data.head()
data.info()
data.describe()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Google Stock Price Over Time')
plt.show()
```

```python
# Volume over time
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Google Stock Volume Over Time')
plt.show()

# Daily price change distribution
data['Daily Change'] = data['Close'] - data['Open']
sns.histplot(data['Daily Change'])
plt.xlabel('Daily Change')
plt.ylabel('Frequency')
plt.title('Distribution of Daily Price Change')
plt.show()

# Correlation between features
correlation_matrix = data[['Open', 'High', 'Low', 'Close', 'Volume']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```
